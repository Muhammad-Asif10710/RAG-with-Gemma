```python
import numpy as np

def query_mistral_rag(user_query, k=1, distance_threshold=1.0):  # lower = more similar
    retriever = db.as_retriever(search_kwargs={"k": k})

    # Get embedding and convert to NumPy
    query_embedding = embedding_model.embed_query(user_query)
    query_embedding_np = np.array([query_embedding])  # shape = (1, dim)

    # FAISS similarity search (returns distances)
    faiss_index = db.index
    D, I = faiss_index.search(query_embedding_np, k)

    # Filter based on distance threshold and retrieve documents
    relevant_docs = []
    for idx, dist in zip(I[0], D[0]):
        if idx != -1 and dist <= distance_threshold:
            # Instead of using docstore.similarity_search_by_vector, retrieve document from FAISS directly
            retrieved_doc = db.similarity_search_by_vector(query_embedding, k=1)[0]  # Use FAISS's similarity search
            relevant_docs.append(retrieved_doc)

    # Build the prompt
    if not relevant_docs:
        prompt = f"""
        You are a Quranic translator and linguistic assistant. A user asked:

        "{user_query}"

        Please respond based on your knowledge, as no relevant ayahs were found in the database.
        """
    else:
        context_block = "\n\n".join([
            f"Ayah Context: {doc.page_content}\nTranslation: {doc.metadata['response']}"
            for doc in relevant_docs
        ])
        prompt = f"""
        You are a Quranic translator and linguistic assistant. A user asked:

        "{user_query}"

        Use the following relevant ayahs and their explanations to generate your response.

        {context_block}

        Answer concisely and accurately.
        """

    # Get the response from the LLM
    output = llm(prompt)[0]["generated_text"]
    return output

# Test query
print(query_mistral_rag("translate to english وَلَا حَرَّمْنَا مِن شَىْءٍۢ ۚ كَذَٰلِكَ كَذَّبَ ٱلَّذِينَ مِن قَبْلِهِمْ حَتَّىٰ ذَاقُوا۟ بَأْسَنَا ۗ قُلْ هَلْ عِندَكُم مِّنْ عِلْمٍۢ فَتُخْرِجُوهُ لَنَآ ۖ إِن تَتَّبِعُونَ إِلَّا ٱلظَّنَّ وَإِنْ أَنتُمْ إِلَّا تَخْرُصُو"))
```

```python
import time
import numpy as np

def measure_similarity_search_time(user_query, k=1, distance_threshold=1.0):  # lower = more similar
    # Get embedding and convert to NumPy
    query_embedding = embedding_model.embed_query(user_query)
    query_embedding_np = np.array([query_embedding])  # shape = (1, dim)

    # Measure the time before performing the similarity search
    start_time = time.time()

    # FAISS similarity search (returns distances)
    faiss_index = db.index
    D, I = faiss_index.search(query_embedding_np, k)

    # Measure the time after performing the similarity search
    end_time = time.time()

    # Calculate the time taken for the similarity search
    time_taken = end_time - start_time
    print(f"Time taken for FAISS similarity search: {time_taken:.4f} seconds")

    # Retrieve the first relevant document based on the FAISS search results
    relevant_docs = []
    if I[0][0] != -1 and D[0][0] <= distance_threshold:  # Check the first document (k=1)
        # Instead of trying to access db._documents, which doesn't exist for FAISS,
        # we can use similarity_search_by_vector which will internally use the
        # FAISS index to retrieve the document.
        retrieved_doc = db.similarity_search_by_vector(query_embedding, k=1)[0]
        relevant_docs.append(retrieved_doc)

    # Return the relevant document and the time taken
    return relevant_docs, time_taken

# Test query and measure time
relevant_docs, time_taken = measure_similarity_search_time("وَلَا حَرَّمْنَا مِن شَىْءٍۢ ۚ كَذَٰلِكَ كَذَّبَ ٱلَّذِينَ مِن قَبْلِهِمْ حَتَّىٰ ذَاقُوا۟ بَأْسَنَا ۗ قُلْ هَلْ عِندَكُم مِّنْ عِلْمٍۢ فَتُخْرِجُوهُ لَنَآ ۖ إِن تَتَّبِعُونَ إِلَّا ٱلظَّنَّ وَإِنْ")

# Print the relevant document and the time taken
print(f"Relevant document: {relevant_docs}")
print(f"Time taken: {time_taken:.4f} seconds")
```
